```python
import pyodbc
import pandas as pd
import matplotlib.pyplot as plt
```

```python
conn = pyodbc.connect('Driver={SQL Server};'
                      'Server=LAPTOP-BRYN\SERVER1;'
                      'Database=AdventureWorks2019;'
                      'Trusted_Connection=yes;')
```

```python
cursor = conn.cursor()
```

```python
query = 'SELECT * FROM Sales.TerritoryPerformance ORDER BY SalesYear, SalesQuarter, TerritoryID'
```

```python
TerritoryP = pd.read_sql(query, conn)
```

```python
print(TerritoryP.head(10))
```

```python


# Calculate the average GrossProfit for each CountryRegionCode per year
avg_profit_by_region = TerritoryP.groupby(['CountryRegionCode', 'SalesYear'])['GrossProfit'].mean()

# Reshape the data so each year is a column (for side-by-side bars)
avg_profit_by_region = avg_profit_by_region.unstack()

# Plot the results
avg_profit_by_region.plot(kind='bar', figsize=(12, 6), colormap='tab10')

# Make the y axis logarithmic so we can see the lower profits
plt.yscale('log')

# Add labels and title
plt.xlabel('CountryRegionCode')
plt.ylabel('Average Gross Profit')
plt.title('Average Gross Profit by Country Region (2013-2014)')
plt.legend(title="Year")  # Add legend to distinguish years

# Show the plot
plt.show()
```

```python
# Calculate the average GrossProfit for each CountryRegionCode per year
avg_sales_by_region = TerritoryP.groupby(['CountryRegionCode', 'SalesYear'])['Sales'].mean()

# Reshape the data so each year is a column (for side-by-side bars)
avg_sales_by_region = avg_sales_by_region.unstack()

# Plot the results
avg_sales_by_region.plot(kind='bar', figsize=(12, 6), colormap='tab10')

# Make the y axis logarithmic so we can see the lower profits
plt.yscale('log')

# Add labels and title
plt.xlabel('CountryRegionCode')
plt.ylabel('Average Sales')
plt.title('Average Sales by Country Region (2013-2014)')
plt.legend(title="Year")  # Add legend to distinguish years

# Show the plot
plt.show()
```
